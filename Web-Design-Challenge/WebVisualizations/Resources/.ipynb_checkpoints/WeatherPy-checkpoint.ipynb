```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy import stats

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "../output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# config information
url = 'http://api.openweathermap.org/data/2.5/weather?'
units = 'imperial'
req_url = url + 'appid=' + weather_api_key + '&units=' + units + '&q='

#creating new list of city names so following lists will have same length
city_names = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
maxt = []
wspeed = []

set_no = 1
attempts = 0
max_att = 50 #per example output

print('Beginning Data Retrieval')
print('-----------------------------')

for city in cities:
    if attempts == max_att:
        time.sleep(60)
        set_no += 1
        attempts = 0
        
    print(f'Processing Record {attempts} of Set {set_no} | {city}')
    
    try:
        ow_response = requests.get(req_url + city).json()
        city_names.append(ow_response['name'])
        cloudiness.append(ow_response['clouds']['all'])
        country.append(ow_response['sys']['country'])
        date.append(ow_response['dt'])
        humidity.append(ow_response['main']['humidity'])
        lat.append(ow_response['coord']['lat'])
        lng.append(ow_response['coord']['lon'])
        maxt.append(ow_response['main']['temp_max'])
        wspeed.append(ow_response['wind']['speed'])
    
    except:
        print('City not found. Skipping...')
        
    attempts += 1
    
print('-----------------------------')
print('Data Retrieval Complete')      
print('-----------------------------')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#length check
#print(len(city_names), len(cloudiness), len(country), len(date), len(humidity), len(lat), len(lng), len(maxt), len(wspeed))

#create dataframe from lists and print count
#added city_names to above loop for list lengths to match

df_dict = {
    'City': city_names,
    'Cloudiness': cloudiness,
    'Country': country,
    'Date': date,
    'Humidity': humidity,
    'Lat': lat,
    'Lng': lng,
    'Max Temp': maxt,
    'Wind Speed': wspeed }

weather_df = pd.DataFrame(df_dict)

weather_df.count()
```

```python
#export to csv and print header to match example output
weather_df.to_csv(output_data_file)

weather_df.head()
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere DataFrames
north_df = weather_df.loc[weather_df['Lat'] > 0]

south_df = weather_df.loc[weather_df['Lat'] < 0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#new vals for north plot
x_vals = north_df['Lat']
y_vals = north_df['Max Temp']

#Max Temp x Lat linear reg
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_vals, y_vals)

reg_vals = x_vals * slope + intercept

eq_str = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))

plt.scatter(x_vals, y_vals)
plt.plot(x_vals,reg_vals,'r-')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.annotate(eq_str,(5,5),fontsize=15,color='red')

print(f"The r-squared is: {rvalue**2}")

plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#new vals for south plot
x_vals = south_df['Lat']
y_vals = south_df['Max Temp']

#Max Temp x Lat linear reg
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_vals, y_vals)

reg_vals = x_vals * slope + intercept

eq_str = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))

plt.scatter(x_vals, y_vals)
plt.plot(x_vals,reg_vals,'r-')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.annotate(eq_str,(-30,50),fontsize=15,color='red')

print(f"The r-squared is: {rvalue**2}")

plt.show()

#Linear regression of Max Temps and Latitude shows this is the strongest relationship studied. 
#With above r-squared values, it holds that latitude is a good indicator of max temp.
```
